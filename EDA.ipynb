```python
df_copy["Size"] = df_copy['Size'].str.replace("M", "")
df_copy["Size"] = df_copy['Size'].str.replace("k", "")
df_copy["Size"] = df_copy['Size'].str.replace("+", "")
df_copy["Size"] = df_copy['Size'].str.replace(",", "")
```

```python
df_copy["Size"] = df_copy['Size'].str.replace("Varies with device", str(np.nan))
```

```python
df_copy[df_copy['Size'] == '1,000+']
```

```python
df_copy["Size"] = df_copy["Size"].astype('float')
```

```python
df_copy["Size"].dtype
```

```python
df_copy.drop(df_copy[~df_copy["Reviews"].str.isnumeric()].index[0], inplace=True)
```

```python
df_copy["Reviews"] = df_copy["Reviews"].astype("int")
```

```python
df_copy["Installs"] = df_copy["Installs"].str.replace("+","")
df_copy["Installs"] = df_copy["Installs"].str.replace(",","")
```

```python
df_copy["Installs"] = df_copy["Installs"].astype("int")
```

```python
df_copy["Type"].value_counts()
```

```python
df_copy["Type"].fillna("Free", inplace=True)
```

```python
df_copy["Price"] = df_copy["Price"].str.replace("$","")
```

```python
df_copy["Price"] = df_copy["Price"].astype("float")
```
